The data used in this application originates from Excel files (5308.xls and 5329.xls). The CSV files (cell_5308.csv and cell_5329.csv) were created by extracting specific columns from these Excel files

```python
import pandas as pd
```

```python
# Reading 5308.xls file
df = pd.read_excel('5308.xls',sheet_name='Detail_67_3_5')
```

```python
df
```

```python
df.drop(["Status","JumpTo","Cycle","Step","Relative Time(h:min:s.ms)"],inplace=True,axis=1)
```

```python
df.head()
```

```python
df_temp = pd.read_excel('5308.xls',sheet_name='DetailTemp_67_3_5')
```

```python
df_temp=df_temp["Auxiliary channel TU1 T(Â°C)"]
```

```python
df_temp.head()
```

```python
df_final = pd.concat([df,df_temp],axis=1)
```

```python
df_final.head()
```

```python
final_data = {
    "Record Index":df_final["Record Index"],
    "Current":df_final["Cur(mA)"],
    "Voltage":df_final["Voltage(V)"],
    "Capacity":df_final["CapaCity(mAh)"],
    "Temperature":df_final["Auxiliary channel TU1 T(Â°C)"],
    "Time":df_final["Absolute Time"], 
}
```

```python
save_to_csv = pd.DataFrame(final_data)
```

```python
save_to_csv.to_csv("cell_5308.csv",index=False)
```
